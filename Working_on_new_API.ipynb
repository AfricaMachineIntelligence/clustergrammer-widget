# Working on new API
Import libraries

```python
import numpy as np
import pandas as pd
from clustergrammer_widget import clustergrammer_widget as cgm_widget
from clustergrammer_widget import Network
```

```python
# generate random matrix
num_rows = 500
num_cols = 10
np.random.seed(seed=100)
mat = np.random.rand(num_rows, num_cols)

# make row and col labels
rows = range(num_rows)
cols = range(num_cols)
rows = [str(i) for i in rows]
cols = [str(i) for i in cols]

# make dataframe 
df = pd.DataFrame(data=mat, columns=cols, index=rows)
```

# old way with variable

```python
net = Network()
net.load_df(df)
net.make_clust()
cgm_widget(network=net.widget())
```

# New API

```python
cgm = cgm_widget()
cgm.load_df(df)
cgm.cluster()
```
